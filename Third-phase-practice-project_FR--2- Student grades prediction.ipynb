```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/FlipRoboTechnologies/ML-Datasets/main/Grades/Grades.csv")
```

```python
df
```

```python
# Check the columns in the dataset
print(df.columns)

```

```python
# Drop irrelevant columns like 'Seat No' if it exists
if 'Seat No' in df.columns:
    df = df.drop(columns=['Seat No'])
```

```python
# Convert non-numeric columns to numeric or drop them
non_numeric_columns = df.select_dtypes(exclude=['int', 'float']).columns
df = df.drop(columns=non_numeric_columns)
```

```python
# Splitting into input features (grades) and target variable (CGPA)
X = df.drop(columns=['CGPA'])  # Input features
y = df['CGPA']  # Target variable
```

```python
# Handle missing values if any
X.fillna(0, inplace=True)  # Fill missing values with 0 for simplicity
```
